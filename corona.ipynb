Import libraries:

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt
from gluonts.model.deepar import DeepAREstimator
from gluonts.dataset.util import to_pandas
from gluonts.trainer import Trainer
from gluonts.dataset.common import ListDataset
```

Read data:

```python
url = 'https://covid.ourworldindata.org/data/ecdc/total_cases.csv'
df = pd.read_csv(url, header=0, index_col=0)
df_new = pd.DataFrame(df.sum(axis = 1, skipna = True), columns=['corona_cases']) 
df_new.head(7)
```

Prepare train set:

```python
training_data = ListDataset([{'start': df_new.index[0], 'target': df_new.corona_cases[:'22.02.2020']}], freq = 'D')
```

Train DeepAR:

```python
estimator = DeepAREstimator(freq='D', prediction_length=30, trainer=Trainer(epochs=100))
predictor = estimator.train(training_data=training_data)
```

Fit-predict:

```python
test_data = ListDataset([{'start': df_new.index[0], 'target': df_new.corona_cases[:'22.02.2020']}], freq = 'D')

for test_entry, forecast in zip(test_data, predictor.predict(test_data)):
    to_pandas(test_entry).plot(linewidth=2, figsize=(10,6), label="historical values")
    forecast.plot(color='g', prediction_intervals=[50.0, 90.0], label="forecast", show_mean=True)

plt.axvline(x= df_new.index[-1], color='gray', linestyle='--', linewidth=6)    
plt.title('Corona cases, worldwide')
plt.legend(loc='upper left')
plt.grid(axis='x')
```
